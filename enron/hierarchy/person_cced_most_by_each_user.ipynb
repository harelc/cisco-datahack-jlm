```python
# Read the data

import pandas as pd
df = pd.read_csv('output.txt', names=['path', 'when', 'id', 'sndr', 'rcpt', 'totype'])
```

```python
# Only interested in these three columns...

df=df[['sndr' ,'rcpt','totype']]
#df
```

```python
# Only interested in message from @enron.com to @enron.com, and only in totype 'Cc'

dfcc = df[df.totype=='Cc'][df.sndr.str.endswith('@enron.com') & df.rcpt.str.endswith('@enron.com')]
# Now we can throw totype away...
df2 = dfcc[['sndr','rcpt']]
#df2.head()
```

```python
# Add a column 'count' -- that's where the count will appear as a result of the groupby.
# after doing the grouping and aggregation, we rest the index to get a flat file again
# -- now of how many times each rcpt was cc'ed by each sndr (not really "each x each",
# only non-0 counts -- we haven't created any new rows...)

df2['count']=1
df3 = df2.groupby(['sndr', 'rcpt']).count()
df4 = df3.reset_index()
#df4.head()
```

```python
# Finally, what we wanted: we can now group by sndr again, and aggregate by ['count'].max
# But since we want the entire row that matches, we need to get the index of the max
# (so idxmax()) and then use that index to access the original, ungrouped, rows...

df5=df4.groupby('sndr')
maxcc_per_sndr = df4.ix[df5['count'].idxmax()]
maxcc_per_sndr
```
